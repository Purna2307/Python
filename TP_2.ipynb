```python
import random
import matplotlib.pyplot as plt

# 1. Création d' un tuple nommé « modalites » pour stocker les modalités.

modalites = ('Voiture', 'Transport en commun', 'Vélo', 'Moto', 'Trottinette', 'À pied')

# 2. Génération d'un échantillon aléatoire de 100 individus

echantillon = random.choices(modalites, weights=[20, 40, 10, 5, 10, 15], k=100)


#3 Création d'un dictionnaire « stat » pour stocker les statistiques 

stat = {}

for elem in modalites:
    stat[elem] = []

for elem in modalites:
    effectif = echantillon.count(elem)
    frequence = effectif / 100
    frequence_pourcentage = frequence * 100
    
    stat[elem].append(effectif)
    stat[elem].append(frequence)
    stat[elem].append(frequence_pourcentage)

#4 Affichage du tableau des statistiques 

print(f"{'Modalité':<20} |{'Effectif':<10}|{'Fréquence':<15} |{'Pourcentage (%)':<20}")
print("-" * 65)
for modalite, valeurs in stat.items():
    effectif, frequence, frequence_pourcentage = valeurs
    print(f"{modalite:<20} |{effectif:<10} |{frequence:<15.2f}| {frequence_pourcentage:<20.2f}")

#5 Recherche du moyen de transport le plus utilisé

modalite_max = None
effectif_max = 0

for modalite, valeurs in stat.items():
    effectif = valeurs[0]
    if effectif > effectif_max:
        effectif_max = effectif
        modalite_max  = modalite

print(f"\nLe moyen de transport le plus utilisé est : {modalite_max} avec un effectif de {effectif_max}.\n")

#6 Tracer un diagramme en bandes

effectifs_list = [valeurs[0] for valeurs in stat.values()]

plt.figure(figsize=(10, 7))  
plt.bar(modalites, effectifs_list, color='blue')  
plt.title('Diagramme des effectifs des moyens de transport')  
plt.ylabel('Les effectifs ')  
plt.xticks(range(len(modalites)), modalites)  
plt.show()  

#7 Tracer un diagramme en Circulaire

print("\n")
effectifs_list = [valeurs[0] for valeurs in stat.values()]
plt.figure(figsize=(8, 4))  
plt.pie(effectifs_list, labels=modalites, autopct='%1.1f%%')  
plt.title('Diagramme des effectifs des moyens de transport')  
plt.show()  
```

```python
import matplotlib.pyplot as plt

genres = ('Action', 'Comédie', 'Romance', 'Drame', 'Science-fiction')
pourcentages = [45.0, 16.0, 8.0, 8.8, 22.2]
total_etudiants = 500

stat = {}

for elem in genres:
    stat[elem] = []

# 2) Calcule et affichage du tableau statistique

for elem, pourcentage in zip(genres, pourcentages):
    effectif = int(pourcentage * total_etudiants / 100)  
    frequence = effectif / total_etudiants  
    frequence_pourcentage = frequence * 100  
    
    stat[elem].append(effectif)
    stat[elem].append(frequence)
    stat[elem].append(frequence_pourcentage)

print(f"{'Genres':<20} |{'Effectif':<10}|{'Fréquence':<15} |{'Pourcentage (%)':<20}")
print("-" * 65)
for modalite, valeurs in stat.items():
    effectif, frequence, frequence_pourcentage = valeurs
    print(f"{modalite:<20} |{effectif:<10} |{frequence:<15.2f}| {frequence_pourcentage:<20.2f}")
print("\n")
    
# 3) Diagramme en bar 

effectifs_list = [valeurs[0] for valeurs in stat.values()] 

plt.figure(figsize=(10, 7))  
plt.bar(genres, effectifs_list, color='red')  
plt.title('Diagramme des effectifs des genres de films préférés')  
plt.ylabel('Effectifs')  
plt.xticks(range(len(genres)), genres)  
plt.show() 

```

```python
import random
import matplotlib.pyplot as plt

#1)

# Création d'un tuple nommé « satisfaction » 
satisfaction = ('Très insatisfait', 'Insatisfait', 'Indifférent', 'Satisfait', 'Très satisfait')

#  Génération d'un échantillon aléatoire de 200 individus
echantillon = random.choices(satisfaction, weights=[3, 10, 20, 50, 17], k=200)

#2)

# Création d'un dictionnaire « stat_satisfaction » pour stocker les statistiques 
stat_satisfaction = {}

for elem in satisfaction:
    stat_satisfaction[elem] = []

for elem in satisfaction:
    effectif = echantillon.count(elem)
    frequence = effectif / 200
    frequence_pourcentage = frequence * 100
    
    stat_satisfaction[elem].append(effectif)
    stat_satisfaction[elem].append(frequence)
    stat_satisfaction[elem].append(frequence_pourcentage)

# Affichage du tableau des statistiques 
print(f"{'Réponse':<20} |{'Effectif':<10}|{'Fréquence':<15} |{'Pourcentage (%)':<20}")
print("-" * 65)
for reponse, valeurs in stat_satisfaction.items():
    effectif, frequence, frequence_pourcentage = valeurs
    print(f"{reponse:<20} |{effectif:<10} |{frequence:<15.2f}| {frequence_pourcentage:<20.2f}")


# 5) 

# Moyenne pondérée 

scores = {
    'Très insatisfait': 1,
    'Insatisfait': 2,
    'Indifférent': 3,
    'Satisfait': 4,
    'Très satisfait': 5
}

# Calcul de la moyenne
total_score = sum([scores[reponse] * valeurs[0] for reponse, valeurs in stat_satisfaction.items()])
moyenne = total_score / 200

# Calcul de la médiane 
effectifs_totaux = sum([valeurs[0] for valeurs in stat_satisfaction.values()])
cumul = 0
for reponse, valeurs in stat_satisfaction.items():
    cumul += valeurs[0]
    if cumul >= effectifs_totaux / 2:
        mediane = reponse
        break
        
# Recherche de la réponse la plus fréquente 
reponse_max = None
effectif_max = 0

for reponse, valeurs in stat_satisfaction.items():
    effectif = valeurs[0]
    if effectif > effectif_max:
        effectif_max = effectif
        reponse_max = reponse

# Affichage des résultats des mesures statistiques

print(f"\nMoyenne de satisfaction : {moyenne:.2f}")
print(f"Médiane de satisfaction : {mediane}")
print(f"La réponse la plus fréquente est : {reponse_max} avec un effectif de {effectif_max}.\n")

#4)

# Tracer un diagramme en Bande

effectifs_list_satisfaction = [valeurs[0] for valeurs in stat_satisfaction.values()]

plt.figure(figsize=(10, 7))  
plt.bar(satisfaction, effectifs_list_satisfaction, color='green')  
plt.title('Diagramme des effectifs de satisfaction des employés')  
plt.ylabel('Les effectifs')  
plt.xticks(range(len(satisfaction)), satisfaction)  
plt.show()  

# Tracer un diagramme en Circulaire

print("\n")
plt.figure(figsize=(8, 4))  
plt.pie(effectifs_list_satisfaction, labels=satisfaction, autopct='%1.1f%%')  
plt.title('Diagramme des effectifs de satisfaction des employés')  
plt.show()
```
